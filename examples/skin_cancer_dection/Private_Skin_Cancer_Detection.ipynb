# Private Skin Cancer Detection

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from utils import preprocess_image

import tf_encrypted as tfe 
from tf_encrypted.private_model import load_graph, secure_model
```

## 1. Pre-process Image

```python
skin_cancer_input = preprocess_image("ISIC_0012836.jpg")
```

## 2. Secure the Model 

```python
with tfe.protocol.SecureNN():
    model_s = secure_model("/tmp/model.pb")
```

## 3. Make Private Prediction

```python
private_logit = model_s.private_predict(skin_cancer_input)
```

```python
if np.argmax(private_logit) == 0:
    print("It's a Melanoma")
else:
    print("It's not a Melanoma")
```
